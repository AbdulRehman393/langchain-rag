## Prompting OpenAI models

```python
from langchain_openai import ChatOpenAI

# Define the LLM
llm = ChatOpenAI(model="gpt-4o-mini", api_key="<OPENAI_API_TOKEN>")

prompt = 'Three reasons for using LangChain for LLM application development.'
response = llm.invoke(prompt)

print(response.content)
```

## 1. ChatOpenAI(...)
Creates the language model.
## 2. llm.invoke(prompt)
Asks the LLM to generate a response to your prompt.
## 3. response.content
Extracts only the text from the modelâ€™s full response object.
## Additional parameters:
'max_completion_tokens', 'temperature', etc

## Prompting Hugging Face models

```python
# Import the HuggingFacePipeline class for defining Hugging Face pipelines
from langchain_huggingface import HuggingFacePipeline

# Define the LLM from the Hugging Face model ID
llm = HuggingFacePipeline.from_model_id(
    model_id="crumb/nano-mistral",
    task="text-generation",
    pipeline_kwargs={"max_new_tokens": 20}
)

prompt = "Hugging Face is"

# Invoke the model
response = llm.invoke(prompt)
print(response)
```
